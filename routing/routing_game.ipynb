```python
"""""
Nonatomic routing game model solved for the user equilibrium (UE) via the Franke-Wolfe algorithm

"""""

def create_graph(rows, cols):
    G = nx.grid_2d_graph(rows, cols)
    
    # assign edge costs
    for edge in G.edges:
        G[edge[0]][edge[1]]['weight'] = 1 # random.uniform(1,5) # random edge weights
        
    return G

def reset_flow(G): 
    for edge in G.edges: 
        G[edge[0]][edge[1]]['weight'] = 1
    return G

def plot_graph(G, rows, cols, paths): 
    
    pos = dict((node, node) for node in G.nodes)
    nx.draw(G, pos, with_labels=False, node_size=300, node_color='skyblue', font_size=8, 
            font_color='black', font_weight='bold', width=2)
    
    path = paths[0]
    nx.draw_networkx_nodes(G, pos, nodelist=path, node_color='darkblue', node_size=300)
    edges = [(path[i], path[i+1]) for i in range(len(path)-1)]
    nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color='darkblue', width=2)
    
    #edge_labels = {(edge[0], edge[1]): G[edge[0]][edge[1]]['weight'] for edge in G.edges}
    #nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    plt.show()
    
def add_agents(G, num_agents): 
    source_sink_dict = {i: {'source': random.choice(list(G.nodes)),
                       'sink': random.choice(list(G.nodes))}
                    for i in range(1, num_agents + 1)}
    
    return source_sink_dict

def congestion_fcn(x_hat, beta=0.5): 
    """""
    linear congestion function depending on flow x_hat
    
    Note: not currently using this function
    
    """""
    fft = 1
    beta = 0.5
    for i in x_hat: 
        cost = fft * (1 + (beta * x_hat[i]))
        
    return cost

def updateTravelTime(G, paths, x_bar):
    """""
    This method updates the travel time on the edges with the current flow from traffic patterns
    
    """""
    fft = 1
    beta = 0.5
    
    for edge in G.edges:
        # congestion fcn
        cost = fft * (1 + beta * x_bar[edge])
        G.edges[edge]['weight'] = cost
           
    return G
    
def dijkstra(G, source, target, weight):
    """""
    calculates shortest path from a specified origin to destination
    
    """""
    path = nx.shortest_path(G, source=source, target=target, weight='weight')
    return path
    
def loadAON(G, source_sink_dict, num_agents): 
    """""
    computes all-or-nothing for optimal flows
    
    """""
    x_opt = {edge: 0 for edge in G.edges}
    SPTT = 0.0
    
    paths = []
    total_cost = []
    
    G = reset_flow(G) 
    
    for agent in range(1, num_agents + 1): 
        source = source_sink_dict[agent]['source']
        sink = source_sink_dict[agent]['sink']
        path = dijkstra(G, source=source, target=sink, weight='weight')
        path_cost = sum(G[path[i]][path[i+1]]['weight'] 
                         for i in range(len(path)-1))
        paths.append(path)
        total_cost.append(path_cost)
        # if agent traverses edge add 1 
        for i in range(len(path) - 1): 
            try:
                edge = (path[i], path[i + 1])
                x_opt[edge] += 1
            except:
                edge = (path[i + 1], path[i])
                x_opt[edge] += 1
    
    SPTT = sum(total_cost)
   
    #print(list(x_bar.values()))
    return SPTT, x_opt, paths

def initial_flow(G, source_sink_dict, num_agents): 
    """""
    compute initial shortest paths on a probabilistic weighted graph
    
    """""
    x_bar = {edge: 0 for edge in G.edges}
    
    # assign random edge costs
    for edge in G.edges:
        G[edge[0]][edge[1]]['weight'] = round(random.uniform(1,5), 3)
    
    paths = []
    total_cost = []
        
    for agent in range(1, num_agents + 1): 
        source = source_sink_dict[agent]['source']
        sink = source_sink_dict[agent]['sink']
        path = dijkstra(G, source=source, target=sink, weight='weight')
        path_cost = sum(G[path[i]][path[i+1]]['weight'] 
                         for i in range(len(path)-1))
        paths.append(path)
        total_cost.append(path_cost)
        # if agent traverses edge add 1 
        for i in range(len(path) - 1): 
            try:
                edge = (path[i], path[i + 1])
                x_bar[edge] += 1
            except:
                edge = (path[i + 1], path[i])
                x_bar[edge] += 1
    
    total_cost = sum(total_cost)
   
    return total_cost, x_bar, paths

def compute_shortest_paths(G, source_sink_dict, num_agents): 
    """""
    re-compute shortest paths to update traffic patterns
    
    """""
    x_hat = {edge: 0 for edge in G.edges}
    
    paths = []
    total_cost = []
        
    for agent in range(1, num_agents + 1): 
        source = source_sink_dict[agent]['source']
        sink = source_sink_dict[agent]['sink']
        path = dijkstra(G, source=source, target=sink, weight='weight')
        path_cost = sum(G[path[i]][path[i+1]]['weight'] 
                         for i in range(len(path)-1))
        paths.append(path)
        total_cost.append(path_cost)
        # if agent traverses edge add 1 
        for i in range(len(path) - 1): 
            try:
                edge = (path[i], path[i + 1])
                x_hat[edge] += 1
            except:
                edge = (path[i + 1], path[i])
                x_hat[edge] += 1
  
    
    total_cost = sum(total_cost)

    return total_cost, x_hat, paths

def findAlpha(k): 
    """""
    Implement line search to compute optimal alpha for FW step size
    
    """""
    # use common step size
    alpha = 2 / (k + 2)
    
    return alpha
        

def franke_wolfe(G, source_sink_dict, num_agents): 
    
    rows = 10
    cols = 10
    accuracy = 0.001
    max_iter = 100
    max_time = 60 #seconds
    x_hat = {edge: 0 for edge in G.edges}
     
    k_iter = 1 # iterations
    gap = float('inf')
    TSTT = float('inf')
    
    # compute optimal shortest paths
    SPTT, x_opt, paths_optimal = loadAON(G, source_sink_dict, num_agents)
    SPTT = round(SPTT, 5) 
    
    # compute inital shortest paths based on probabilistic weights
    total_cost_initial, x_bar, paths = initial_flow(G, source_sink_dict, num_agents)
    
    startTime = time.time()
    
    while gap > accuracy and k_iter <= max_iter:
        print(f"\n\nIteration {k_iter}: ", "Gap Value: ", gap)
        
        # determine Franke-Wolfe step size alpha
        alpha = findAlpha(k_iter)
        
        # update flow
        for edge in G.edges: 
            # update x_bar based on updated flow from x_hat
            x_bar[edge] = round((1 - alpha) * x_bar[edge] + alpha * x_hat[edge], 4)
        
        print(x_bar.values())
        
        # Compute new travel times
        G = updateTravelTime(G, paths, x_bar)
        
        # re-compute shortest paths
        total_cost_updated, x_hat, paths = compute_shortest_paths(G, source_sink_dict, num_agents)
        
        print("Total Cost: ", round(total_cost_updated, 3))
        
        
        for edge in G.edges: 
            TSTT = 0.0 
            flow = np.array(list(x_hat.values())) 
            cost = G[edge[0]][edge[1]]['weight']
            TSTT += flow * cost
        TSTT = sum(TSTT)
        TSTT = round(TSTT, 5)
        
        
        #gap = (TSTT/SPTT) - 1
        gap = round((total_cost_updated/SPTT) - 1, 4)
        #print("TSTT:, ", TSTT) 
        print("SPTT:, ", SPTT)
        print("Gap: ", gap)
        
        plot_graph(G, rows, cols, paths)
             
        if gap < 0: 
            print("Error gap is less than 0") 
         
        k_iter += 1
        
        if k_iter > max_iter:
            print("The algorithm did not converge to a desired gap in the max number of iterations")
            print("Algorithm took: ", round(time.time() - startTime, 4), "Seconds")
            print("Current gap: ", round(gap, 4)) 
                
        if time.time() - startTime > max_time:  
            print("The algorithm did not converge to a desired gap in the alloted max time")
            print("Algorithm went through: ", max_iter, "iterations")
            print("Current gap: ", round(gap, 4))
                

def main():
    
    rows = 10
    cols = 10
    num_agents = 8
    
    G  = create_graph(rows, cols) 
    source_sink_dict = add_agents(G, num_agents)

    franke_wolfe(G, source_sink_dict, num_agents) 
      
    #plot_graph(G, rows, cols, paths)
    
if __name__ == '__main__': 
    main()
```
